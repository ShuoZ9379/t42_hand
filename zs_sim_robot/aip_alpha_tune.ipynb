```python
from common import plot_util as pu
import pickle,os,copy
import numpy as np
from matplotlib import pyplot as plt

x=np.arange(0,1.01,0.01)
y_ln=x
y_sq=x**2
y_sqrt=np.sqrt(x)
y_ln_mid=copy.copy(x)
y_sq_mid=copy.copy(x)
y_sqrt_mid=copy.copy(x)
for i in range(len(x)):
    if x[i]<1/3:
        y_ln_mid[i]=0
        y_sq_mid[i]=0
        y_sqrt_mid[i]=0
    elif x[i]>2/3:
        y_ln_mid[i]=1
        y_sq_mid[i]=1
        y_sqrt_mid[i]=1
    else:
        y_ln_mid[i]=3*x[i]-1
        y_sq_mid[i]=(3*x[i]-1)**2
        y_sqrt_mid[i]=np.sqrt(3*x[i]-1)
plt.plot(x,y_ln)
plt.xlabel('Training Progress')
plt.ylabel('Probability of forcing alpha=0')
plt.show()

plt.plot(x,y_sq)
plt.xlabel('Training Progress')
plt.ylabel('Probability of forcing alpha=0')
plt.show()

plt.plot(x,y_sqrt)
plt.xlabel('Training Progress')
plt.ylabel('Probability of forcing alpha=0')
plt.show()

plt.plot(x,y_ln_mid)
plt.xlabel('Training Progress')
plt.ylabel('Probability of forcing alpha=0')
plt.show()

plt.plot(x,y_sq_mid)
plt.xlabel('Training Progress')
plt.ylabel('Probability of forcing alpha=0')
plt.show()

plt.plot(x,y_sqrt_mid)
plt.xlabel('Training Progress')
plt.ylabel('Probability of forcing alpha=0')
plt.show()
```
